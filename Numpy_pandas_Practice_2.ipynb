Q1. Employee Salary Adjustment
A company stores salaries in a NumPy array.

salaries = np.array([45000, 52000, 61000, 48000, 75000])

Tasks:
1.

*  Increase all salaries by 10%.
* Find the average salary
*   Find the maximum and minimum salary
*  Print salaries greater than 60000.


```python
import numpy as np

salaries = np.array([45000, 52000, 61000, 48000, 75000])

# 1. Increase all salaries by 10%
updated_salaries = salaries * 1.10
print("Updated Salaries:", updated_salaries)

# 2. Average salary
print("Average Salary:", np.mean(updated_salaries))

# 3. Max and Min salary
print("Max Salary:", np.max(updated_salaries))
print("Min Salary:", np.min(updated_salaries))

# 4. Salaries greater than 60000
print("Salaries > 60000:", updated_salaries[updated_salaries > 60000])
```

Q2. Sales Data Analysis

Daily sales of a store for 10 days:

sales = np.array([120,150,180,200,170,160,210,220,190,175])

Tasks:

* Compute total sales.


* Compute mean and standard deviation.

* Identify days where sales exceeded the average.

* Normalize the sales using Min-Max scaling.

```python
sales = np.array([120,150,180,200,170,160,210,220,190,175])

# 1. Total sales
print("Total Sales:", np.sum(sales))

# 2. Mean and Standard Deviation
print("Mean:", np.mean(sales))
print("Std Dev:", np.std(sales))

# 3. Days where sales exceeded average
avg = np.mean(sales)
print("Above Average Sales:", sales[sales > avg])

# 4. Min-Max Normalization
normalized = (sales - np.min(sales)) / (np.max(sales) - np.min(sales))
print("Normalized Sales:", normalized)
```

Q4. Employee Dataset Exploration

Create a DataFrame:

data = {
'Employee':['A','B','C','D','E','F'],

'Department':['HR','IT','IT','Finance','HR','Finance'],

'Salary':[50000,65000,70000,72000,52000,68000],

'Experience':[2,5,7,6,3,4]
}

Tasks:
1. Convert this dictionary into a DataFrame.
2. Display basic statistics of the dataset.
3. Find the average salary per department.
4. Identify employees with experience greater than 4 years.
5. Add a new column Bonus = 10% of salary.

```python
import pandas as pd

data = {
    'Employee':['A','B','C','D','E','F'],
    'Department':['HR','IT','IT','Finance','HR','Finance'],
    'Salary':[50000,65000,70000,72000,52000,68000],
    'Experience':[2,5,7,6,3,4]
}

df = pd.DataFrame(data)
end="\n"+100*'_'+"\n"
# 1. DataFrame
print("1. DataFrame",end,df)

# 2. Statistics
print("2. Statistics",end,df.describe())

# 3. Average salary per department
print("3. Average salary per department",end,df.groupby('Department')['Salary'].mean())

# 4. Experience > 4
print("4. Experience > 4",end,df[df['Experience'] > 4])

# 5. Bonus column
df['Bonus'] = df['Salary'] * 0.10
print("5.New Column Bonus added",end,df)
```

Q5. Missing Data Handling

Dataset:
data = {
'Name':['John','Anna','Peter','Linda','David'],
'Age':[28, np.nan, 35, 32, np.nan],
'Salary':[50000,54000,np.nan,58000,60000]
}

Tasks:
1. Detect missing values.
2. Replace missing Age with mean age.
3. Replace missing Salary with median salary.
4. Remove rows with any remaining missing values.

```python
data = {
    'Name':['John','Anna','Peter','Linda','David'],
    'Age':[28, np.nan, 35, 32, np.nan],
    'Salary':[50000,54000,np.nan,58000,60000]
}

df = pd.DataFrame(data)

# 1. Detect missing values
print("1. Detect missing values",end,df.isnull().sum())

# 2. Fill Age with mean
df['Age'] = df['Age'].fillna(int(df['Age'].mean()))
print("Fill Age with mean",end,df)

# 3. Fill Salary with median
df['Salary'] = df['Salary'].fillna(df['Salary'].median())
print("salary with median",end,df)

# 4. Drop remaining missing
df.dropna(inplace=True)

print("Drop remaining missing values",end,df)
```

Q7. Customer Purchase Analysis

Dataset:

data = {
'CustomerID':[101,102,103,104,105,106,107],
'Age':[25,45,35,50,23,40,60],
'PurchaseAmount':[250,600,400,800,150,700,900],
'City':['Delhi','Mumbai','Delhi','Chennai','Mumbai','Delhi','Chennai']
}

Tasks:
1. Find the average purchase amount by city.
2. Identify customers with purchase above overall average.
3. Find the city generating the highest revenue.
4. Create a column Age_Group:

    o <30 → Young

    o 30–50 → Adult
    
    o 50 → Senior
5. Find average purchase per age group.

```python
data = {
    'CustomerID':[101,102,103,104,105,106,107],
    'Age':[25,45,35,50,23,40,60],
    'PurchaseAmount':[250,600,400,800,150,700,900],
    'City':['Delhi','Mumbai','Delhi','Chennai','Mumbai','Delhi','Chennai']
}

df = pd.DataFrame(data)

# 1. Avg purchase by city
print(df.groupby('City')['PurchaseAmount'].mean())

```

```python
# 2. Above average customers
avg = df['PurchaseAmount'].mean()
print(df[df['PurchaseAmount'] > avg])

```

```python
# 3. Highest revenue city
print(df.groupby('City')['PurchaseAmount'].sum().idxmax())

```

```python
# 4. Age group
def age_group(age):
    if age < 30:
        return "Young"
    elif age <= 50:
        return "Adult"
    else:
        return "Senior"

df['Age_Group'] = df['Age'].apply(age_group)
df
```

```python

# 5. Avg purchase per age group
print(df.groupby('Age_Group')['PurchaseAmount'].mean())
```

Q8. Time Series Sales Analysis

Generate data:

dates = pd.date_range('2024-01-01', periods=12, freq='M')

sales = [200,220,250,270,300,320,350,360,380,400,420,450]

df = pd.DataFrame({'Date':dates,'Sales':sales})

Tasks:
1. Convert Date to datetime index.
2. Compute monthly percentage growth.
3. Calculate 3-month rolling average sales.
4. Identify the month with highest growth rate.

```python
dates = pd.date_range('2024-01-01', periods=12, freq='ME')
sales = [200,220,250,270,300,320,350,360,380,400,420,450]

df = pd.DataFrame({'Date':dates,'Sales':sales})

# 1. Set datetime index
df.set_index('Date', inplace=True)
df
```

```python

# 2. Monthly growth
df['Growth'] = df['Sales'].pct_change() * 100
df
```

```python
# 3. Rolling average
df['Rolling_Avg'] = df['Sales'].rolling(window=3).mean()

df
```

```python
# 4. Highest growth month
print(df['Growth'].idxmax())

```
